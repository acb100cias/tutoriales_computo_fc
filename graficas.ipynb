```python
from numpy import *
from matplotlib.pyplot import *
```

```python
X=arange(-4,4,0.01)
Y=arange(0.1,4,0.01)
Z=arange(0.1,4,0.1)
```

```python
f1=figure()
plot(sin(X),label='sen(x)')
plot(cos(X),label='cos(x)')
title('Sen y Cos')
legend()
#savefig('senocoseno.pdf')
#show()
```

```python
f2=figure()
plot(exp(Y),label='epx(x)')
plot(log(Y),label='ln(x)')
title('Exp y Log')
legend()
#savefig('expolog1.pdf')
#show()
```

```python

f3=figure()
plot(exp(Z),label='exp(x)',color='blue',marker='o')
plot(log(Z),label='ln(x)',color='blue',marker='*')
title('Exp y Log')
legend()
savefig('expolog2.pdf')
#show()
```

```python
X=arange(0.1,5,0.01)
f4=figure()
for i in arange(0.01,1,0.1):
    plot(i*exp(0.2*X))
```

```python
def logistica(t,xmax,x0,a):
    return ((x0*xmax/(xmax-x0))*exp(a*t))/(1+(x0/(xmax-x0)*exp(a*t)))
```

```python
X=arange(0,10,0.1)
for i in arange(0.01,1,0.1):
    plot(logistica(X,100.,60.,i))
    plot(logistica(X,100.,45.,i))
    plot(logistica(X,100.,120.,i))
```
