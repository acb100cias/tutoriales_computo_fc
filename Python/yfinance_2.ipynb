```python
tickers=['AMZN','NFLX','GOOG']
tik=[yf.Ticker(ticker)for ticker in tickers]
```

```python
tik
```

```python
DF=[]
for t in tik:
    fin=t.financials
    bs=t.balancesheet
    cf=t.cashflow
    T=pd.concat([fin,bs,cf])
    D=T.T
    D=D.reset_index()
    D.columns=['Fecha',*D.columns[1:]]
    D['Ticker']=t.ticker
    DF.append(D)
    
    
```

```python
DF
```

```python
T2=pd.concat(DF, ignore_index=True)
T2=T2.set_index(['Ticker','Fecha'])
```
