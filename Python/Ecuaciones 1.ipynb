## Problema 1

$$\dfrac{\mathrm{d}x}{\mathrm{d}t}=\alpha x(t)$$

## Problema 2

$$
\dfrac{\mathrm{d}x}{\mathrm{d}t}=\alpha x(1-\dfrac{x}{k})
$$

```python
from numpy import *
from matplotlib.pyplot import *
from scipy.integrate import *
```

```python
a=0.01
T=arange(1,800,0.1)
x0=0.1
```

```python
def f(x,t):
    return a*x
```

```python
sol1=odeint(f,x0,T)
```

```python
plot(T,sol1)
```

```python
for ci in arange(0.1,2,0.1):
    sol1=odeint(f,ci,T)
    plot(T,sol1)
```

```python
k=20
def g(x,t):
    return a*x*(1-(x/k))-x**2/k
```

```python
sol2=odeint(g,10,T)
```

```python
plot(T,sol2)
```

```python
for ci in arange(2,40,1):
    sol2=odeint(g,ci,T)
    plot(T,sol2)
```
