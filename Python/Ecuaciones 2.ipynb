## Problema 3. Sistema de Lorenz


$$
\dot{x}=\sigma(y-x)\\
\dot{y}=x(\rho-z)-y\\
\dot{z}=xy-\beta z
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def f(Y,t):
    y1,y2=Y
    return [y2,-k*y1-b*y2+4*np.cos(2*t)]
```

```python
k=10.
b=2.
```

```python
y0=[0.5,0]
```

```python
T=np.arange(0,20,0.1)
ys=odeint(f,y0,T)
```

```python
ys
```

```python
plt.plot(T,ys[:,0],label='posici√≥n')
plt.plot(T,ys[:,1],label='Velocidad')
plt.legend()
```

```python
plt.plot(ys[:,0],ys[:,1])
```

```python
T=np.arange(0,50,0.1)
for ci in [0,0.5,1,1.5,2,2.5]:
    y0=[ci,0]
    ys=odeint(f,y0,T)
    plt.plot(ys[:,0],ys[:,1],label='%f'%ci)
plt.legend()
plt.title('Trayectorias en el plano fase')
plt.xlabel('Posici√≥n')
plt.ylabel('velocidad')
```

```python
def lorenz(x, y, z, s=10, r=28, b=2.667):
    x_dot = s*(y - x)
    y_dot = r*x - y - x*z
    z_dot = x*y - b*z
    return [x_dot, y_dot, z_dot]
```

```python
sol_lorenz=odeint(lorenz(0,1.,1.05),[0,1,1.05],np.arange(0,100,0.1))
```

```python
dt = 0.01
num_steps = 10000

xs = np.empty(num_steps + 1)
ys = np.empty(num_steps + 1)
zs = np.empty(num_steps + 1)


xs[0], ys[0], zs[0] = (0., 1., 1.05)
```

```python

for i in range(num_steps):
    x_dot, y_dot, z_dot = lorenz(xs[i], ys[i], zs[i])
    xs[i + 1] = xs[i] + (x_dot * dt)
    ys[i + 1] = ys[i] + (y_dot * dt)
    zs[i + 1] = zs[i] + (z_dot * dt)


    
```

```python
plt.plot(xs)
plt.plot(ys)

plt.plot(zs)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(xs, ys, zs, lw=0.5)
ax.set_title("Atractor de Lorenz")
```
