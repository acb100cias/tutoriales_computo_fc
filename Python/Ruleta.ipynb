```python
import random
```

```python
class Roulette():
    def __init__(self):
        self.pockets=[]
        for i in range(1,37):
            self.pockets.append(i)
        self.ball= None
        self.pocketOdds = len(self.pockets)-1
    def spin(self):
        self.ball = random.choice(self.pockets)
    def betPocket(self,pocket,mnt):
        if str(pocket) == str(self.ball):
            return mnt*self.pocketOdds
        else: return -mnt
    def __str__(self):
        return 'Ruleta'    
        
```

```python
def playRoulette(game,N,pocket,bet):
    totPocket=0
    for i in range(N):
        game.spin()
    totPocket += game.betPocket(pocket,bet)
    if totPocket:
        print(N, 'vueltas de', game)
        print('Pago esperado ', pocket,'=',str(100*totPocket/N)+'%')
    return(totPocket/N)
    
```

```python

game=Roulette()

```

```python
for N in (100,1000000):
    for i in range(1000):
        playRoulette(game,N,5,1)
    
```
