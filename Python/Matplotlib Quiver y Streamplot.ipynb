```python
fig, (ax1,ax2,ax3) = subplots(1,3,figsize=(9,9))


# first subplot
x = arange(0,2.2,0.2)
y = arange(0,2.2,0.2)
X, Y = meshgrid(x, y)
u = cos(X)*Y
v = sin(y)*Y
n = -2
color = sqrt(((v-n)/2)*2 + ((u-n)/2)*2)


ax1.quiver(X,Y,u,v,color, alpha=0.8)


ax1.xaxis.set_ticks([])
ax1.yaxis.set_ticks([])
ax1.axis([-0.2, 2.3, -0.2, 2.3])
ax1.set_aspect('equal')
ax1.set_title('meshgrid function')


# second subplot
x = arange(-2,2.2,0.2)
y = arange(-2,2.2,0.2)
X, Y = meshgrid(x, y)
z = X*exp(-X**2 -Y**2)
dx, dy = gradient(z)
n = -2
color = sqrt(((dx-n)/2)*2 + ((dy-n)/2)*2) 


ax2.quiver(X,Y,dx,dy,color)


ax2.xaxis.set_ticks([])
ax2.yaxis.set_ticks([])
ax2.set_aspect('equal')
ax2.set_title('gradient')


# third subplot
x = arange(0,2*pi+2*pi/20,2*pi/20)
y = arange(0,2*pi+2*pi/20,2*pi/20)
X,Y = meshgrid(x,y)
u = sin(X)*cos(Y)
v = -cos(X)*sin(Y)
n = -1
color = sqrt(((dx-n)/2)*2 + ((dy-n)/2)*2)


ax3.quiver(X,Y,u,v,color)


ax3.xaxis.set_ticks([])
ax3.yaxis.set_ticks([])
ax3.axis([0,2*pi,0,2*pi])
ax3.set_aspect('equal')
ax3.set_title('four vortices')


# save and show the figure
tight_layout()
fig.savefig('3_quiver_plots.png', dpi=300, bbox_inches='tight')


show()
```

```python
x = arange(0,10)
y = arange(0,10)


X, Y = meshgrid(x,y)
u = ones((10,10)) # x-component to the right
v = zeros((10,10)) # y-component zero


fig, ax = subplots()


ax.streamplot(X,Y,u,v, density = 0.5)
ax.set_title('Stream Plot of Parallel Lines')
show()
```

```python
x = arange(-4,4,0.2)
y = arange(-4,4,0.2)


X,Y = meshgrid(x,y)
Ex = (X + 1)/((X+1)**2 + Y**2) - (X - 1)/((X-1)**2 + Y**2)
Ey = Y/((X+1)**2 + Y**2) - Y/((X-1)**2 + Y**2)


fig, ax = subplots(figsize=(9,9))


ax.streamplot(X,Y,Ex,Ey,density=1.5)


ax.set_aspect('equal')
ax.plot(-1,0,'-or')
ax.plot(1,0,'-ob')
ax.set_title('Stream Plot of Two Point Charges')


show()
```

```python
from matplotlib.patches import Circle
  
# Function to determin electric field
def E(q, r0, x, y):
    den = hypot(x-r0[0], y-r0[1])**3
    return q * (x - r0[0]) / den, q * (y - r0[1]) / den
  
# Grid of x, y points
nx, ny = 64, 64
x = linspace(-2, 2, nx)
y = linspace(-2, 2, ny)
X, Y = meshgrid(x, y)
  
# Create a multipole with nq charges of
# alternating sign, equally spaced
# on the unit circle.
  
# Increase the power with increase in charge
nq = 2**1 
charges = []
for i in range(nq):
    q = i % 2 * 2 - 1
    charges.append((q, (np.cos(2 * np.pi * i / nq),
                        np.sin(2 * np.pi * i / nq))))
  
# Electric field vector, E =(Ex, Ey)
# as separate components
Ex, Ey = zeros((ny, nx)), zeros((ny, nx))
  
for charge in charges:
    ex, ey = E(*charge, x = X, y = Y)
    Ex += ex
    Ey += ey
  
fig = figure(figsize =(18, 8))
ax = fig.add_subplot(111)
  
# Plotting the streamlines with 
# proper color and arrow
color = 2 * log(hypot(Ex, Ey))
ax.streamplot(x, y, Ex, Ey, color = color,
              linewidth = 1, cmap = cm.inferno,
              density = 2, arrowstyle ='->', 
              arrowsize = 1.5)
  
# Add filled circles for the charges
# themselves
charge_colors = {True: '#AA0000', 
                 False: '#0000AA'}
  
for q, pos in charges:
    ax.add_artist(Circle(pos, 0.05, 
                         color = charge_colors[q>0]))
  
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_aspect('equal')
  
show()
```
