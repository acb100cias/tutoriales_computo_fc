```python
from pandas import *
from wordcloud import *
import matplotlib.pyplot as plt
```

```python
vyt=read_csv('/home/augusto/Descargas/data kaggle/videos-stats.csv')
```

```python
vyt=vyt.drop('Unnamed: 0',axis=1)
```

```python
vyt
```

```python
vytc=read_csv('/home/augusto/Descargas/data kaggle/comments.csv')
```

```python
vytc=vytc.drop('Unnamed: 0',axis=1)
```

```python
vytc
```

```python
D1=vyt.dropna().filter(['Video ID','Keyword'])
```

```python
D2=vytc.filter(['Video ID','Comment'])
```

```python
D3=D1.merge(D2,on='Video ID')
```

```python
DataFrame(D3.Keyword.values).drop_duplicates()[0]
```

```python
D4=D3.loc[D3.Keyword=='data science']
```

```python
DataFrame(D4.groupby('Video ID'))
```

```python
texto=''
for i in D4.Comment.values:
    texto+=i
```

```python
texto
```

```python
stopwords = set(STOPWORDS)
```

```python
wc=WordCloud(width=1200, height=1800, stopwords=stopwords,max_words=200,background_color='white').generate_from_text(texto)
```

```python
plt.imshow(wc)
plt.axis('off')
```
