```python
from numpy import *
from matplotlib.pyplot import *
import seaborn as sb
```

```python
X=random.normal(loc=5,size=100000)
```

```python
sb.displot(X)
```

```python
X1=random.binomial(n=10,p=0.5,size=100000)
sb.displot(X1)
```

```python
sb.distplot(X1)
sb.distplot(X)
```

```python
X2=random.poisson(lam=2,size=100000)
sb.distplot(X2)
```

```python
Y1=random.binomial(n=1000,p=0.01,size=100000)
Y2=random.poisson(lam=10,size=100000)
sb.distplot(Y1)
sb.distplot(Y2)
```

```python
Z1=random.normal(scale=2,size=100000)
Z2=random.logistic(size=100000)
sb.distplot(Z1)
sb.distplot(Z2)
```
