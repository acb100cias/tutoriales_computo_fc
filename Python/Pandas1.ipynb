```python
import pandas as pd
```

```python
dic={'autos':['VW','Nissan','Ford'],'color':['rojo','azul','negro']}
```

```python
autos_posibles=['VW','Nissan','Ford','Renault', 'Toyota','chevrolet']
colores_posibles=['rojo','azul','negro','amarillo','plata', 'blanco']
```

```python
import numpy as np
A=[]
C=[]
for i in range(500):
    A.append(np.random.choice(autos_posibles))
    C.append(np.random.choice(colores_posibles))
    
```

```python
dic2={'autos':A,'color':C}
```

```python
DataCoches=pd.DataFrame(dic2)
```

```python
v=[]
for i in range(500):
    v.append(np.random.choice(range(10,121)))
```

```python
DataCoches['velocidad']=v
```

```python
DataCoches
```

```python
DataCoches.groupby('color').describe()
```

```python
DataCoches.pivot_table(index='autos',columns='color',aggfunc='count',margins=True,margins_name='Total').to_csv('tablaCoches.csv')
```

```python
DataCoches.pivot_table(index='autos',columns='color',aggfunc='count',margins=True,margins_name='Total').to_excel('tablaCoches.xlsx')
```
