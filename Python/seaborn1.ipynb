```python
from pandas import *
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
sns.relplot(data=tips,x="total_bill",y="tip",col="time",hue="smoker",size="size",style="sex")
```

```python
dots=sns.load_dataset("dots")
```

```python
sns.relplot(data=dots,kind="line",x="time",y="firing_rate",hue="choice",col="align",size="coherence")
```

```python
dots
```

```python
fmri=sns.load_dataset("fmri")
```

```python
sns.relplot(data=fmri,kind="line",x='timepoint',y='signal',col='region',hue='event',style='event')
```

```python
sns.lmplot(data=tips, x='total_bill',y='tip',col='time',hue='smoker')
```

```python
sns.catplot(data=tips,kind="swarm",x='day',y='total_bill',hue='sex')
```

```python
sns.catplot(data=tips,kind="violin",x='day',y='tip',hue='sex',split=True,col='time')
```

```python
penguins=sns.load_dataset("penguins")
```

```python
sns.relplot(data=penguins,x='bill_length_mm',y="bill_depth_mm",hue='species',col='sex',size='body_mass_g')
```

```python
sns.pairplot(data=penguins,hue='species')
```
