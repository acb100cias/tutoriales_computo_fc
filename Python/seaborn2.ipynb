```python
from numpy import *
from pandas import *
import seaborn as sns
```

```python
vuelos=sns.load_dataset('flights')
```

```python
vuelos
```

```python
sns.relplot(data=vuelos,x='year',y='passengers',hue='month',kind='line')
```

```python
vuelos2=vuelos.pivot(index='year',columns='month',values='passengers')
```

```python
vuelos2
```

```python
sns.relplot(data=vuelos2,kind='line')
```

```python
sns.relplot(data=vuelos,x='month',y='passengers',hue='year',kind='line')
```

```python
sns.relplot(data=vuelos2.T,kind='line')
```

```python
sns.catplot(data=vuelos2,kind='box')
```

```python
vuelos_dict=vuelos.to_dict()
```

```python
sns.relplot(data=vuelos_dict,x='year',y='passengers',hue='month',kind='line')
```

```python
vuelos_prom=vuelos.groupby('year').mean()
```

```python
sns.relplot(data=vuelos_prom,x='year',y='passengers',kind='line')
```

```python
year=vuelos_prom.index
pasajeros=vuelos_prom.passengers
#sns.relplot(x=)
```

```python
sns.relplot(x=year.to_numpy(),y=pasajeros.to_list(),kind='line')
```
