```python
import numpy as np
```

```python
N=1000000
n_ci=0
n_cu=0
```

```python
l_x=[]
l_y=[]
```

```python
for _ in range(N):
    x=np.random.uniform(-1,1)
    l_x.append(x)
    y=np.random.uniform(-1,1)
    l_y.append(y)
```

```python
l_x
```

```python
l_y
```

```python
for i in range(N):
    normC=l_x[i]**2+l_y[i]**2
    if normC <= 1:
        n_ci+=1
    n_cu+=1
```

```python
import matplotlib.pyplot as plt

for i in range(N):
    normC=l_x[i]**2+l_y[i]**2
    if normC<=1:
        plt.plot(l_x[i],l_y[i],'o',color='blue')
    else:
        plt.plot(l_x[i],l_y[i],'o',color='red')
```
