```python
from pandas import *
from numpy import *
import seaborn as sns
```

```python
prop=sns.load_dataset('tips')
```

```python
prop
```

```python
sns.regplot(data=prop, x='total_bill',y='tip')
```

```python
sns.lmplot(data=prop, x='total_bill',y='tip')
```

```python
sns.lmplot(data=prop,x='size',y='tip')
```

```python
sns.lmplot(data=prop,x='size',y='tip',x_jitter=.05)
```

```python
sns.lmplot(data=prop,x='size',y='tip',x_estimator=mean)
```

```python
sns.lmplot(x='x',y='y',data=anscombe.query("dataset == 'I'"))
```

```python
sns.lmplot(x='x',y='y',data=anscombe.query("dataset == 'II'"))
```

```python
sns.lmplot(data=prop,x='total_bill',y='gprop',y_jitter=0.03,logistic=True)
```

```python
sns.lmplot(data=prop,x='total_bill',y='tip',y_jitter=0.03,lowess=True,line_kws={'color':'red'})
```
