```python
from PIL import Image
from urllib.request import urlopen
import matplotlib.pyplot as plt 
```

```python
url='https://sites.northwestern.edu/kovacslab/files/2019/05/network-550x310.png'
img=Image.open(urlopen(url))
```

```python
img.show()
```

```python
plt.imshow(img)
plt.axis('off')
```

```python
wm_img=img.copy()
```

```python
from PIL import ImageDraw
from PIL import ImageFont

dr=ImageDraw.Draw(wm_img)
```

```python
w,h=img.size
x,y = int(w/2),int(h/2)
```

```python
if x> y:
    fsize=y
elif y>x:
    fsize=x
else:
    fsize = x
```

```python
font = ImageFont.truetype("/usr/share/fonts/truetype/freefont/FreeMono.ttf",int(fsize/4))
```

```python
dr.text((x,y),'Gris es'+'\n'+'toda teoría',fill=150,font=font)
```
